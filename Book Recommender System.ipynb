## Popularity Based Recommender System 

```python
rating.head()
```

```python
book.head()
```

```python
rat_name=rating.merge(book,on='ISBN')
```

```python
rat_name.shape
```

```python
rat_name.head()
```

```python
num_rating=rat_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':'num-rating'},inplace=True)
```

```python
num_rating
```

```python
avg_rating=rat_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'avg-rating'},inplace=True)
avg_rating
```

```python
popular_df=num_rating.merge(avg_rating,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num-rating']>=250].sort_values('avg-rating',ascending=False).head(50)
```

```python
popular_df
```

```python
popular_df.merge(book,on='Book-Title')
```

```python
popular_df=popular_df.merge(book,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num-rating','avg-rating']]
```

## Collaborative Filtering Based Recommender System 

```python
x=rat_name.groupby('User-ID').count()['Book-Rating']>200
smart_user=x[x].index
```

```python
filtered=rat_name[rat_name['User-ID'].isin(smart_user)]
```

```python
y=filtered.groupby('Book-Title').count()['Book-Rating']>50
famous_books=y[y].index
```

```python
Final_ratings=filtered[filtered['Book-Title'].isin(famous_books)]
```

```python
Final_ratings
```

```python
pt=Final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.shape
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    #index fetch
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    for i in similar_items:
        print(pt.index[i[0]])
        
    #return suggestion
```
